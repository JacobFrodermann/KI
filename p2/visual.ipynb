```python
import matplotlib.pyplot as plt
from IPython.display import display

graphs = []
current_y = []

with open("log") as f:
    for line in f:
        line = line.strip()
        if not line:
            continue
        if line == "===":
            if current_y:
                graphs.append(current_y)
            current_y = []
            continue

        if line.startswith('['):
            continue 
        else:
            current_y.append(float(line))

if current_y:
    graphs.append(current_y)

fig, ax = plt.subplots()
for i, y_values in enumerate(graphs, 1):
    ax.plot(range(len(y_values)), y_values)
    ax.set_title(f"Log {i}")
    ax.set_xlabel("Index")
    ax.set_ylabel("Fitness")
    #display(fig)
    #plt.close(fig)
```
